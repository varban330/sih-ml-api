```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(threshold=np.inf)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

## Electricity 

```python
dataset = pd.read_csv('WIFI/WIFI.csv')
dataset
```

```python
k = [1,2,3,4]
df = dataset.iloc[:, k] 
X = dataset.iloc[:, k].values
Y = dataset.iloc[:, 6].values
len(X[0])
df
```

```python
# Save Encoder Objects for new incoming data
import pickle
encx0 = LabelEncoder().fit(X[:,0])
encx1 = LabelEncoder().fit(X[:,1])
encx2 = LabelEncoder().fit(X[:,2])
encx3 = LabelEncoder().fit(X[:,3])
filename = 'WIFI/WIFIencx0.sav'
pickle.dump(encx0,open(filename,'wb'))
filename = 'WIFI/WIFIencx1.sav'
pickle.dump(encx1,open(filename,'wb'))
filename = 'WIFI/WIFIencx2.sav'
pickle.dump(encx2,open(filename,'wb'))
filename = 'WIFI/WIFIencx3.sav'
pickle.dump(encx3,open(filename,'wb'))
```

```python
data = ['Speed','Lan Wire','No','Available']
```

```python
X[:,0] = encx0.transform(X[:,0])
X[:,1] = encx1.transform(X[:,1])
X[:,2] = encx2.transform(X[:,2])
X[:,3] = encx3.transform(X[:,3])
```

```python
print(X)
```

```python
onehotencoder = OneHotEncoder(categorical_features=[0,1,2,3])
onehenc = onehotencoder.fit(X)
filename = 'WIFI/WIFIonehenc.sav'
pickle.dump(onehenc,open(filename,'wb'))
X = onehenc.transform(X).toarray()
d = encx0.transform([data[0]])
data[0] = d[0]
d = encx1.transform([data[1]])
data[1] = d[0]
d = encx2.transform([data[2]])
data[2] = d[0]
d = encx3.transform([data[3]])
data[3] = d[0]
ar = np.array(data)
print(data)
ar = ar.reshape(1,-1)
kar=onehenc.transform(ar).toarray()
print(kar)
print(len(X[0]))
```

```python
k = list()
notlist = [0,3,7,9]
for i in range(11):
    if i not in notlist:
        k.append(i)
X = X[:,k]
kar = kar[:,k]
```

```python
print((kar[0]))
```

## SVR

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
model1 = SVR(kernel='rbf')
model1.fit(X_train,Y_train)
Y_pred = model1.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[1,10,100,1000], 'kernel': ['rbf'], 'gamma':[0.05,0.01,0.001,0.0001]},
             {'C':[100,200,300,400], 'kernel': ['poly'], 'gamma':[0.05,0.06,0.07,0.08,0.09,0.1]},
             {'C':[1,10,100,1000], 'kernel': ['sigmoid'], 'gamma':[0.05,0.01,0.001,0.0001]}]
gridsearch = GridSearchCV(estimator=model1,
                         param_grid=parameters,
                         scoring='r2',
                         cv=10)
gridsearch = gridsearch.fit(X_train,Y_train)
print(gridsearch.best_score_)
print(gridsearch.best_params_)
```

## Random Forest

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
model2 = RandomForestRegressor(n_estimators=700, criterion='mse',random_state=0)
model2 = model2.fit(X,Y)
```

```python
from sklearn.metrics import r2_score
Y_pred = model2.predict(X)
for i in Y_pred:
    print(round(i))
print(Y_test)
print(r2_score(Y,Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'n_estimators':[1,10,100,1000], 'criterion':['mse','mae']}]
grid_search = GridSearchCV(estimator=model2,
                          param_grid=parameters,
                          scoring='r2',
                          cv=10)
grid_search = grid_search.fit(X,Y)
print(grid_search.best_score_)
print(grid_search.best_params_)
```
