## Scholarship Portal 

```python
dataset = pd.read_csv('scholarship_disbursement/scholarship_disbursement.csv')
dataset
```

```python
k = [1,2,3,4,7,8]
X = dataset.iloc[:, k].values
Y = dataset.iloc[:, 11].values
len(X[0])
```

```python
# Save Encoder Objects for new incoming data
import pickle
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encx1 = LabelEncoder().fit(X[:,1])
encx3 = LabelEncoder().fit(X[:,3])
encx4 = LabelEncoder().fit(X[:,4])
filename = 'scholarship_disbursement/scholarship_disbursementencx1.sav'
pickle.dump(encx1,open(filename,'wb'))
filename = 'scholarship_disbursement/scholarship_disbursementencx3.sav'
pickle.dump(encx3,open(filename,'wb'))
filename = 'scholarship_disbursement/scholarship_disbursementencx4.sav'
pickle.dump(encx4,open(filename,'wb'))
```

```python
data = [8.2,'COE',3,'SBI','DD',2019]
```

```python
X[:,1] = encx1.transform(X[:,1])
X[:,3] = encx3.transform(X[:,3])
X[:,4] = encx4.transform(X[:,4])
print(max(X[:,4]))
onehotencoder = OneHotEncoder(categorical_features=[1,2,3,4,5])
onehenc = onehotencoder.fit(X)
filename = 'scholarship_disbursement/scholarship_disbursementonehenc.sav'
pickle.dump(onehenc,open(filename,'wb'))
X = onehenc.transform(X).toarray()
d = encx1.transform([data[1]])
data[1] = d[0]
d = encx3.transform([data[3]])
data[3] = d[0]
d = encx4.transform([data[4]])
data[4] = d[0]
ar = np.array(data)
print(data)
ar = ar.reshape(1,-1)
kar=onehenc.transform(ar).toarray()
print(kar)
print(len(X[0]))
```

```python
k = list()
notlist = [0,8,12,15,20]
for i in range(23):
    if i not in notlist:
        k.append(i)
X = X[:,k]
kar = kar[:,k]
```

```python
print(len(X[0]))
```

## SVR

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2, random_state = 0)
```

```python
model1 = SVR(kernel='rbf')
model1.fit(X_train,Y_train)
Y_pred = model1.predict(X_test)
```

```python
Y_pred
```

```python
Y_test
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'C':[1,10,100,1000], 'kernel': ['rbf'], 'gamma':[0.05,0.01,0.001,0.0001]},
             {'C':[100,200,300,400], 'kernel': ['poly'], 'gamma':[0.05,0.06,0.07,0.08,0.09,0.1]},
             {'C':[1,10,100,1000], 'kernel': ['sigmoid'], 'gamma':[0.05,0.01,0.001,0.0001]}]
gridsearch = GridSearchCV(estimator=model1,
                         param_grid=parameters,
                         scoring='r2',
                         cv=10)
gridsearch = gridsearch.fit(X_train,Y_train)
print(gridsearch.best_score_)
print(gridsearch.best_params_)
```
